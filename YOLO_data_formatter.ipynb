```python
import json

# Function to read and parse JSON from a file
def parse_json_from_file(file_path):
    with open(file_path, 'r') as file:
        data = json.load(file)
    return data

# Function to save the modified data with YOLO format
def save_json_to_file(data, file_path):
    with open(file_path, 'w') as file:
        json.dump(data, file, indent=4)

# Example file paths (you can replace these with actual file paths when needed)
input_file_path = '/content/final_merged.json'
output_file_path = '/content/final_merged_yolo.json'

# Parse the JSON file
parsed_data = parse_json_from_file(input_file_path)

# Function to convert bbox to YOLO format
def convert_to_yolo_format(bbox, img_width, img_height):
    x_min, y_min, x_max, y_max = bbox
    # Calculate center coordinates (x_center, y_center)
    x_center = (x_min + x_max) / 2.0
    y_center = (y_min + y_max) / 2.0
    # Calculate width and height of the bbox
    bbox_width = x_max - x_min
    bbox_height = y_max - y_min
    # Normalize the values
    x_center /= img_width
    y_center /= img_height
    bbox_width /= img_width
    bbox_height /= img_height
    return [x_center, y_center, bbox_width, bbox_height]

# Convert each bbox to YOLO format and save to the output file
for entry in parsed_data:
    bbox_yolo = convert_to_yolo_format(entry["bbox"], entry["width"], entry["height"])
    entry["bbox_yolo"] = bbox_yolo

# Save the updated JSON data to a new file
save_json_to_file(parsed_data, output_file_path)

parsed_data  # Display the parsed data with YOLO bbox values
```
